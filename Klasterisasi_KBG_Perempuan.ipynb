Menambahkan Kolom Komnas Perempuan, Data Lembaga Mitra, Badilag.
Kemudian hasilnya ditampung di kolom total kbg perempuan

```python
data21['Total KBG Perempuan'] = data21['Komnas Perempuan'] + data21['Data Lembaga Mitra'] + data21['Badilag']
data21
```

```python
data22['Total KBG Perempuan'] = data22['Komnas Perempuan'] + data22['Data Lembaga Mitra'] + data22['Badilag']
data22
```

### Data Training

```python
x_train21 = data21[['Total KBG Perempuan']].values
x_train22 = data22[['Total KBG Perempuan']].values
```

```python
x_train21
```

```python
x_train22
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train21 = scaler.fit_transform(x_train21)
x_train22 = scaler.fit_transform(x_train22)
```

```python
x_train21
```

```python
x_train22
```

### Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans
```

### 2021

```python
y_cluster21 = kmeans.fit_predict(x_train21)
```

```python
y_cluster21
```

```python
data21['Cluster'] = y_cluster21
```

```python
data21
```

```python
kmeans.cluster_centers_
```

### Visualisasi Data Hasil Clustering

```python
plt.figure(figsize=(10, 5))
plt.scatter(x_train21[:, 0], data21['Provinsi'], c=kmeans.labels_, cmap='viridis')
plt.legend()
plt.show()
```

```python
colors = ['red', 'green', 'blue']
plt.figure(figsize=(10, 5))
for i in range(len(colors)):
    plt.scatter(x_train21[kmeans.labels_ == i, 0], data21['Provinsi'][kmeans.labels_ == i], c=colors[i], label=f'Cluster {i}')

# Menambahkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```

### Klaster Terbaik

```python
inertias = []
k_range = range(1,10)
for k in k_range:
  km = KMeans(n_clusters=k).fit(x_train21)
  inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.plot(k_range, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of squared distances (Inertia)')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
inertias22 = []
k_range = range(1,10)
for k in k_range:
  km = KMeans(n_clusters=k).fit(x_train22)
  inertias22.append(km.inertia_)
```

```python
plt.plot(k_range, inertias22, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Sum of squared distances (Inertia)')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score

# Membuat array untuk menyimpan nilai DBI untuk setiap jumlah cluster
dbi_scores = []

# Rentang jumlah cluster yang ingin diuji
k_range = range(2, 10)

for k in k_range:
    # Membuat model KMeans dengan jumlah cluster k
    kmeans_model = KMeans(n_clusters=k, random_state=42)
    kmeans_labels = kmeans_model.fit_predict(x_train21)  # Gantilah x_train21 dengan data Anda

    # Menghitung nilai Davies-Bouldin Index
    dbi = davies_bouldin_score(x_train21, kmeans_labels)  # Gantilah x_train21 dengan data Anda

    # Menyimpan nilai DBI untuk jumlah cluster saat ini
    dbi_scores.append(dbi)

# Menampilkan plot untuk visualisasi
plt.plot(k_range, dbi_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Davies-Bouldin Index')
plt.title('Davies-Bouldin Index For Optimal k')
plt.grid(True)
plt.show()
```
